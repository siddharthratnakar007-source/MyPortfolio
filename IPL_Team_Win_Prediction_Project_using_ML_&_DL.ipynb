```python
df3.drop(["id", "season","city","date", "player_of_match", 'umpire1', "venue", "umpire2"], axis=1, inplace=True)
```

```python
df3
```

```python
df3.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X=df3.drop(["winner"],axis=1)
```

```python
for col in df3.select_dtypes(include='object').columns:
    df3[col] = le.fit_transform(df3[col])
```

```python
df3.dropna(inplace=True)
```

```python
X = df3.drop(["winner"], axis=1)
y = df3["winner"]
```

```python
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.32, random_state=42)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier             # it is a bagging technique
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
le = LogisticRegression()
```

```python
le.fit(X_train,y_train)
```

```python
y_pred = le.predict(X_test)
```

```python
import keras
```

```python
model = keras.Sequential()
```

```python
model.summary()
```

```python
model.add(
    keras.Input(
        shape=(30,),
        batch_size=32
    )
)
```

```python
model.add(                                 #hidden layer
    keras.layers.Dense(
    64,
    activation='relu'
))
```

```python
model.add(                      # hidden layer
    keras.layers.Dense(
    128,
    activation='relu'
))
```

```python
df3['winner'].value_counts()
```

```python
len(df3['winner'].value_counts())
```

```python
model.add(                         ## Output layer
        keras.layers.Dense(
    16,
    activation='softmax'
)
)
```

```python
model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    x=X_train,
    y=y_train,
    batch_size=32,
    epochs=50,
    verbose="auto",
    validation_data=(X_test,y_test),
 )
```

```python
model.evaluate(X_test,y_test)
```
